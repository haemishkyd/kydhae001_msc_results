## Synchronised Camers Run 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
subset_0_data = pd.read_csv("output_folderDICe_solution_0.txt")
subset_1_data = pd.read_csv("output_folderDICe_solution_1.txt")
subset_2_data = pd.read_csv("output_folderDICe_solution_2.txt")
z_displacement_0 = subset_0_data.MODEL_DISPLACEMENT_Z
z_displacement_1 = subset_1_data.MODEL_DISPLACEMENT_Z
z_displacement_2 = subset_2_data.MODEL_DISPLACEMENT_Z
z_displacement_0 = z_displacement_0[10:110]
z_displacement_1 = z_displacement_1[10:110]
z_displacement_2 = z_displacement_2[10:110]
z_displacement_0 *= -1
z_displacement_1 *= -1
z_displacement_2 *= -1

y_displacement_0 = subset_0_data.MODEL_DISPLACEMENT_Y
y_displacement_1 = subset_1_data.MODEL_DISPLACEMENT_Y
y_displacement_2 = subset_2_data.MODEL_DISPLACEMENT_Y
y_displacement_0 = y_displacement_0[10:110]
y_displacement_1 = y_displacement_1[10:110]
y_displacement_2 = y_displacement_2[10:110]

x_displacement_0 = subset_0_data.MODEL_DISPLACEMENT_X
x_displacement_1 = subset_1_data.MODEL_DISPLACEMENT_X
x_displacement_2 = subset_2_data.MODEL_DISPLACEMENT_X
x_displacement_0 = x_displacement_0[10:110]
x_displacement_1 = x_displacement_1[10:110]
x_displacement_2 = x_displacement_2[10:110]
x_displacement_0 *= -1
x_displacement_1 *= -1
x_displacement_2 *= -1
```

```python
plt.title("Z Displacement")
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
plt.ylim(-0.01,0.05)
plt.plot(z_displacement_0)
plt.plot(z_displacement_1)
plt.plot(z_displacement_2)
```

```python
plt.title("Y Displacement")
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
plt.ylim(-0.01,0.05)
plt.plot(y_displacement_0)
plt.plot(y_displacement_1)
plt.plot(y_displacement_2)
```

```python
plt.title("X Displacement")
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
plt.ylim(-0.01,0.05)
plt.plot(x_displacement_0)
plt.plot(x_displacement_1)
plt.plot(x_displacement_2)
```

```python
mean = np.mean(z_displacement_0, axis=0)
sd = np.std(z_displacement_0, axis=0)
final_displacement_z_0 = [x for x in z_displacement_0 if (x > mean - 2 * sd)]
final_displacement_z_0 = [x for x in final_displacement_z_0 if (x < mean + 2 * sd)]
final_displacement_z_0 = pd.Series(final_displacement_z_0)

mean = np.mean(z_displacement_1, axis=0)
sd = np.std(z_displacement_1, axis=0)
final_displacement_z_1 = [x for x in z_displacement_1 if (x > mean - 2 * sd)]
final_displacement_z_1 = [x for x in final_displacement_z_1 if (x < mean + 2 * sd)]
final_displacement_z_1 = pd.Series(final_displacement_z_1)

mean = np.mean(z_displacement_2, axis=0)
sd = np.std(z_displacement_2, axis=0)
final_displacement_z_2 = [x for x in z_displacement_2 if (x > mean - 2 * sd)]
final_displacement_z_2 = [x for x in final_displacement_z_2 if (x < mean + 2 * sd)]
final_displacement_z_2 = pd.Series(final_displacement_z_2)


plt.plot(final_displacement_z_0)
plt.plot(final_displacement_z_1)
plt.plot(final_displacement_z_2)
```

```python
mean = np.mean(y_displacement_0, axis=0)
sd = np.std(y_displacement_0, axis=0)
final_displacement_y_0 = [x for x in y_displacement_0 if (x > mean - 2 * sd)]
final_displacement_y_0 = [x for x in final_displacement_y_0 if (x < mean + 2 * sd)]
final_displacement_y_0 = pd.Series(final_displacement_y_0)

mean = np.mean(y_displacement_1, axis=0)
sd = np.std(y_displacement_1, axis=0)
final_displacement_y_1 = [x for x in y_displacement_1 if (x > mean - 2 * sd)]
final_displacement_y_1 = [x for x in final_displacement_y_1 if (x < mean + 2 * sd)]
final_displacement_y_1 = pd.Series(final_displacement_y_1)

mean = np.mean(y_displacement_2, axis=0)
sd = np.std(y_displacement_2, axis=0)
final_displacement_y_2 = [x for x in y_displacement_2 if (x > mean - 2 * sd)]
final_displacement_y_2 = [x for x in final_displacement_y_2 if (x < mean + 2 * sd)]
final_displacement_y_2 = pd.Series(final_displacement_y_2)


plt.plot(final_displacement_y_0)
plt.plot(final_displacement_y_1)
plt.plot(final_displacement_y_2)
```

```python
mean = np.mean(x_displacement_0, axis=0)
sd = np.std(x_displacement_0, axis=0)
final_displacement_x_0 = [x for x in x_displacement_0 if (x > mean - 2 * sd)]
final_displacement_x_0 = [x for x in final_displacement_x_0 if (x < mean + 2 * sd)]
final_displacement_x_0 = pd.Series(final_displacement_x_0)

mean = np.mean(x_displacement_1, axis=0)
sd = np.std(x_displacement_1, axis=0)
final_displacement_x_1 = [x for x in x_displacement_1 if (x > mean - 2 * sd)]
final_displacement_x_1 = [x for x in final_displacement_x_1 if (x < mean + 2 * sd)]
final_displacement_x_1 = pd.Series(final_displacement_x_1)

mean = np.mean(x_displacement_2, axis=0)
sd = np.std(x_displacement_2, axis=0)
final_displacement_x_2 = [x for x in x_displacement_2 if (x > mean - 2 * sd)]
final_displacement_x_2 = [x for x in final_displacement_x_2 if (x < mean + 2 * sd)]
final_displacement_x_2 = pd.Series(final_displacement_x_2)


plt.plot(final_displacement_x_0)
plt.plot(final_displacement_x_1)
plt.plot(final_displacement_x_2)
```

```python
steady_state_0 = final_displacement_z_0
steady_state_1 = final_displacement_z_1
steady_state_2 = final_displacement_z_2
plt.plot(steady_state_0)
plt.plot(steady_state_1)
plt.plot(steady_state_2)
z_val1 = np.sqrt(((steady_state_0 ** 2) .sum ()/steady_state_0.size))
z_val2 = np.sqrt(((steady_state_1 ** 2) .sum ()/steady_state_1.size))
z_val3 = np.sqrt(((steady_state_2 ** 2) .sum ()/steady_state_2.size))
print("Subset 1 RMS  : {} mm".format(round(z_val1,5)))
print("Subset 2 RMS  : {} mm".format(round(z_val2,5)))
print("Subset 3 RMS  : {} mm".format(round(z_val3,5)))
print ("Avg Z Axis RMS: {} mm".format(round(((z_val1+z_val2+z_val3)/3),5)))
```

```python
steady_state_0 = final_displacement_y_0
steady_state_1 = final_displacement_y_1
steady_state_2 = final_displacement_y_2
plt.plot(steady_state_0)
plt.plot(steady_state_1)
plt.plot(steady_state_2)
y_val1 = np.sqrt(((steady_state_0 ** 2) .sum ()/steady_state_0.size))
y_val2 = np.sqrt(((steady_state_1 ** 2) .sum ()/steady_state_1.size))
y_val3 = np.sqrt(((steady_state_2 ** 2) .sum ()/steady_state_2.size))
print("Subset 1 RMS  : {} mm".format(round(y_val1,5)))
print("Subset 2 RMS  : {} mm".format(round(y_val2,5)))
print("Subset 3 RMS  : {} mm".format(round(y_val3,5)))
print ("Avg Y Axis RMS: {} mm".format(round(((y_val1+y_val2+y_val3)/3),5)))
```

```python
steady_state_0 = final_displacement_x_0
steady_state_1 = final_displacement_x_1
steady_state_2 = final_displacement_x_2
plt.plot(steady_state_0)
plt.plot(steady_state_1)
plt.plot(steady_state_2)
x_val1 = np.sqrt(((steady_state_0 ** 2) .sum ()/steady_state_0.size))
x_val2 = np.sqrt(((steady_state_1 ** 2) .sum ()/steady_state_1.size))
x_val3 = np.sqrt(((steady_state_2 ** 2) .sum ()/steady_state_2.size))
print("Subset 1 RMS  : {} mm".format(round(x_val1,5)))
print("Subset 2 RMS  : {} mm".format(round(x_val2,5)))
print("Subset 3 RMS  : {} mm".format(round(x_val3,5)))
print ("Avg X Axis RMS: {} mm".format(round(((x_val1+x_val2+x_val3)/3),5)))
```
