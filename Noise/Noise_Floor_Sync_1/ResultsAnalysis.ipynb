```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
subset_0_data = pd.read_csv("output_folderDICe_solution_0.txt")
subset_1_data = pd.read_csv("output_folderDICe_solution_1.txt")
subset_2_data = pd.read_csv("output_folderDICe_solution_2.txt")
z_displacement_0 = subset_0_data.MODEL_DISPLACEMENT_Z
z_displacement_1 = subset_1_data.MODEL_DISPLACEMENT_Z
z_displacement_2 = subset_2_data.MODEL_DISPLACEMENT_Z
z_displacement_0 = z_displacement_0[20:120]
z_displacement_1 = z_displacement_1[20:120]
z_displacement_2 = z_displacement_2[20:120]
plt.plot(z_displacement_0)
plt.plot(z_displacement_1)
plt.plot(z_displacement_2)
```

```python
mean = np.mean(z_displacement_0, axis=0)
sd = np.std(z_displacement_0, axis=0)
final_displacement_0 = [x for x in z_displacement_0 if (x > mean - 2 * sd)]
final_displacement_0 = [x for x in final_displacement_0 if (x < mean + 2 * sd)]
final_displacement_0 = pd.Series(final_displacement_0)

mean = np.mean(z_displacement_1, axis=0)
sd = np.std(z_displacement_1, axis=0)
final_displacement_1 = [x for x in z_displacement_1 if (x > mean - 2 * sd)]
final_displacement_1 = [x for x in final_displacement_1 if (x < mean + 2 * sd)]
final_displacement_1 = pd.Series(final_displacement_1)

mean = np.mean(z_displacement_2, axis=0)
sd = np.std(z_displacement_2, axis=0)
final_displacement_2 = [x for x in z_displacement_2 if (x > mean - 2 * sd)]
final_displacement_2 = [x for x in final_displacement_2 if (x < mean + 2 * sd)]
final_displacement_2 = pd.Series(final_displacement_2)


plt.plot(final_displacement_0)
plt.plot(final_displacement_1)
plt.plot(final_displacement_2)
```

```python
steady_state_0 = final_displacement_0
plt.plot(steady_state_0)
np.sqrt(((steady_state_0 ** 2) .sum ()/steady_state_0.size))
```

```python
steady_state_1 = final_displacement_1
plt.plot(steady_state_1)
np.sqrt(((steady_state_1 ** 2) .sum ()/steady_state_1.size))
```

```python
steady_state_2 = final_displacement_2
plt.plot(steady_state_2)
np.sqrt(((steady_state_2 ** 2) .sum ()/steady_state_2.size))
```
