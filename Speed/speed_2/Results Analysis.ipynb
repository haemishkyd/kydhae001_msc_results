```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

files = glob.glob("output_folderDICe*.txt")
final_list = sorted(files)
all_displacements = []
z_displacements = []
y_displacements = []
x_displacements = []

for file_name in final_list:
    data = pd.read_csv(file_name)
    all_displacements.append(data)
    z_displacements.append(data.MODEL_DISPLACEMENT_Z)
    y_displacements.append(data.MODEL_DISPLACEMENT_Y)
    x_displacements.append(data.MODEL_DISPLACEMENT_X)

data_1 = pd.read_csv("timing.txt")
execution_speed_1 = data_1["Entire Process"]
sample_rate = np.average(execution_speed_1)
x_data=np.arange(0,len(z_displacements[0])*sample_rate,sample_rate)
print(x_data)
```

```python
plt.rcParams['figure.figsize'] = [40/2.54, 32/2.54]
plt.title("Z Displacement",fontsize = 30)
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
plt.plot(x_data,z_displacements[0])
for disp in z_displacements[:94]:
    plt.plot(x_data,disp,'o')
```
