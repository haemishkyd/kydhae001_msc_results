```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

files = glob.glob("output_folderDICe*.txt")
final_list = sorted(files)
all_displacements_1 = []
z_displacements_1 = []


for file_name in final_list:
    data = pd.read_csv(file_name)
    all_displacements_1.append(data)
    z_displacements_1.append(data.MODEL_DISPLACEMENT_Z)


data_1 = pd.read_csv("timing.txt")
execution_speed_1 = data_1["Entire Process"]
sample_rate = np.average(execution_speed_1)
x_data_1=np.arange(0,len(z_displacements_1[0])*sample_rate,sample_rate)
```

```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

files = glob.glob("../speed_2/output_folderDICe*.txt")
final_list = sorted(files)
all_displacements_2 = []
z_displacements_2 = []

for file_name in final_list:
    data = pd.read_csv(file_name)
    all_displacements_2.append(data)
    z_displacements_2.append(data.MODEL_DISPLACEMENT_Z)

data_2 = pd.read_csv("timing.txt")
execution_speed_2 = data_2["Entire Process"]
sample_rate = np.average(execution_speed_2)
x_data_2=np.arange(0,len(z_displacements_2[0])*sample_rate,sample_rate)
x_data_2 = x_data_2 -3
```

```python
files = glob.glob("../speed_3/output_folderDICe*.txt")
final_list = sorted(files)
all_displacements_3 = []
z_displacements_3 = []


for file_name in final_list:
    data = pd.read_csv(file_name)
    all_displacements_3.append(data)
    z_displacements_3.append(data.MODEL_DISPLACEMENT_Z)


data_3 = pd.read_csv("timing.txt")
execution_speed_3 = data_1["Entire Process"]
sample_rate = np.average(execution_speed_3)
x_data_3=np.arange(0,len(z_displacements_3[0])*sample_rate,sample_rate)
x_data_3 = x_data_3 +17
```
