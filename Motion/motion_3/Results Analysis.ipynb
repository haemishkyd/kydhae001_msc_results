```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

plt.rcParams.update({'font.size': 20})

files = glob.glob("output_folderDICe*.txt")
final_list = sorted(files)
all_displacements = []
z_displacements = []
y_displacements = []
x_displacements = []

for file_name in final_list:
    data = pd.read_csv(file_name)
    all_displacements.append(data)
    z_displacements.append(data.MODEL_DISPLACEMENT_Z)
    y_displacements.append(data.MODEL_DISPLACEMENT_Y)
    x_displacements.append(data.MODEL_DISPLACEMENT_X)
```

```python
plt.rcParams['figure.figsize'] = [40/2.54, 32/2.54]
plt.title("Z Displacement",fontsize = 30)
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
for disp in z_displacements:
    plt.plot(disp)
```

```python
plt.title("Y Displacement",fontsize = 30)
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
for disp in y_displacements:
    plt.plot(disp)
```

```python
plt.title("X Displacement",fontsize = 30)
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
for disp in x_displacements:
    plt.plot(disp)
```
