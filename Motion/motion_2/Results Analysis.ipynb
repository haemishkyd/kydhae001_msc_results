```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

plt.rcParams.update({'font.size': 20})
files = glob.glob("output_folderDICe*.txt")
final_list = sorted(files)
all_displacements = []
z_displacements = []
y_displacements = []
x_displacements = []

for file_name in final_list:
    data = pd.read_csv(file_name)
    all_displacements.append(data)
    z_displacements.append(data.MODEL_DISPLACEMENT_Z)
    y_displacements.append(data.MODEL_DISPLACEMENT_Y)
    x_displacements.append(data.MODEL_DISPLACEMENT_X)
```

```python
plt.rcParams['figure.figsize'] = [40/2.54, 32/2.54]
plt.title("Z Displacement",fontsize = 30)
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
for disp in z_displacements:
    plt.plot(disp)
```

```python
array_stp1F = np.array([])
array_stp2F = np.array([])
array_stp3F = np.array([])
array_stp4 = np.array([])
array_stp3B = np.array([])
array_stp2B = np.array([])
array_stp1B = np.array([])
for disp in z_displacements:    
    array_stp1F = np.append(array_stp1F, disp[22])
    array_stp2F = np.append(array_stp2F, disp[25])
    array_stp3F = np.append(array_stp3F, disp[28])
    array_stp4 = np.append(array_stp4, disp[31])
    array_stp3B = np.append(array_stp3B, disp[35])
    array_stp2B = np.append(array_stp2B, disp[37])
    array_stp1B = np.append(array_stp1B, disp[40])
array_stp1F = array_stp1F.reshape(15,8)
array_stp2F = array_stp2F.reshape(15,8)
array_stp3F = array_stp3F.reshape(15,8)
array_stp4 = array_stp4.reshape(15,8)
array_stp3B = array_stp3B.reshape(15,8)
array_stp2B = array_stp2B.reshape(15,8)
array_stp1B = array_stp1B.reshape(15,8)
print(array_stp1F)
error_array = np.absolute(array_stp1F-1)
print ("Error: "+str(np.average(error_array)))
print("Variance:"+str(np.std(array_stp1F)))
print(array_stp2F)
error_array = np.absolute(array_stp2F-2)
print ("Error: "+str(np.average(error_array)))
print("Variance:"+str(np.std(array_stp2F)))
print(array_stp3F)
error_array = np.absolute(array_stp3F-3)
print ("Error: "+str(np.average(error_array)))
print("Variance:"+str(np.std(array_stp3F)))
print(array_stp4)
error_array = np.absolute(array_stp4-4)
print ("Error: "+str(np.average(error_array)))
print("Variance:"+str(np.std(array_stp4)))
print(array_stp3B)
error_array = np.absolute(array_stp3B-3)
print ("Error: "+str(np.average(error_array)))
print("Variance:"+str(np.std(array_stp3B)))
print(array_stp2B)
error_array = np.absolute(array_stp2B-2)
print ("Error: "+str(np.average(error_array)))
print("Variance:"+str(np.std(array_stp2B)))
print(array_stp1B)
error_array = np.absolute(array_stp1B-1)
print ("Error: "+str(np.average(error_array)))
print("Variance:"+str(np.std(array_stp1B)))
```
