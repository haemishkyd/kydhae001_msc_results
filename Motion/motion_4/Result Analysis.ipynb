```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

files = glob.glob("output_folderDICe*.txt")
final_list = sorted(files)
all_displacements = []
z_displacements = []
y_displacements = []
x_displacements = []

for file_name in final_list:
    data = pd.read_csv(file_name)
    all_displacements.append(data)
    z_displacements.append(data.MODEL_DISPLACEMENT_Z)
    y_displacements.append(data.MODEL_DISPLACEMENT_Y)
    x_displacements.append(data.MODEL_DISPLACEMENT_X)
    
print(len(z_displacements))
```

```python
plt.rcParams['figure.figsize'] = [40/2.54, 32/2.54]
plt.title("Z Displacement",fontsize = 30)
plt.ylabel("Displacement (mm)")
plt.xlabel("Time (s)")
plt.grid()
for disp in z_displacements[:94]:
    plt.plot(disp)
```

```python
np_array = np.array([])
for disp in z_displacements:    
    np_array = np.append(np_array, disp[25])
np_array = np_array.reshape(12,8)
print(np_array)
error_array = np.absolute(np_array-10)
print ("Error: "+str(np.average(error_array)))
```
