```python
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

data_1 = pd.read_csv("timing.txt")
execution_speed_1 = data_1["Entire Process"]

data_2 = pd.read_csv("../subsets_2/timing.txt")
execution_speed_2 = data_2["Entire Process"]

data_3 = pd.read_csv("../subsets_3/timing.txt")
execution_speed_3 = data_3["Entire Process"]

data_4 = pd.read_csv("../subsets_4/timing.txt")
execution_speed_4 = data_4["Entire Process"]
```

```python
results=[]
results.append(np.average(execution_speed_1))
results.append(np.average(execution_speed_2))
results.append(np.average(execution_speed_3))
results.append(np.average(execution_speed_4))
print(results)
x=[60,30,15,7]
```

```python
plt.rcParams['figure.figsize'] = [40/2.54, 32/2.54]
plt.title("Subset Processing Time", fontsize=20)
plt.ylabel("Average Processing Time(s)")
plt.xlabel("Size of Subsets (pixels)")
plt.vlines(x=[60,30,15,7], ymin=0, ymax=2,colors='purple',linestyles="dashed")
plt.grid()
plt.plot(x,results,marker='o')
```

```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

files = glob.glob("output_folderDICe*.txt")
final_list = sorted(files)
all_displacements = []
z_displacements = []
y_displacements = []
x_displacements = []

print (len(final_list))

for file_name in final_list:
    data = pd.read_csv(file_name)
    all_displacements.append(data)
    z_displacements.append(data.MODEL_DISPLACEMENT_Z)
    y_displacements.append(data.MODEL_DISPLACEMENT_Y)
    x_displacements.append(data.MODEL_DISPLACEMENT_X)
```
